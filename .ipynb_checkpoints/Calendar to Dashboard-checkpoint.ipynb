```python
import pandas as pd 
import numpy as np
import glob 
import os

def read_files(directory):
    files = glob.glob(os.path.join(directory, '*.csv'))
    new_df = pd.DataFrame()
    for file in files:
        df = pd.read_csv(file)
        file_name = os.path.splitext(os.path.basename(file))[0].split('_')[0]
        df["Type"] = file_name
        new_df = pd.concat([new_df, df])
        new_df = new_df.sort_values(by = ["Start"])
    return new_df
```

```python
read_files("Files")
```
